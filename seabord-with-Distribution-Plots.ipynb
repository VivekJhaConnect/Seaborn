```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

#### Import Data

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

### Distribution Plots

##### Distribution Plot

```python
sns.histplot(crash_df['not_distracted'], kde=True)
```

```python
sns.histplot(crash_df['not_distracted'], kde=False)
```

```python
sns.histplot(crash_df['not_distracted'], kde=False, bins=25)
```

### Joint Plots

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pair Plots

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

### Rug Plot

```python
sns.rugplot(tips_df['tip'])
```
