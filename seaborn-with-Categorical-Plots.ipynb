```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

#### Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.var)
```

#### Count Plot

```python
sns.countplot(x ='sex', data=tips_df)
```

#### Box Plot

```python
sns.boxenplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc=0)
```

#### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

#### Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data=tips_df)
```

```python

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True)
```

```python

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex')
```

```python

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
```

#### Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df)
```

#### Palettes

```python
plt.figure(figsize=(8,5))

sns.set_style('dark')

sns.set_context('talk')

sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
```
